```python
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('dark_background')
```

```python
import numpy as np
import scipy.io
```

```python
import Miscellaneous as misc
```

```python
mat = scipy.io.loadmat('MarksData.mat')
simU = mat['UMatrix']
simJ = mat['JMatrix']
depth = mat['potential'][:,0]
sep = mat['separation'][:,0]
```

```python
JoU = np.zeros(simU.shape)
```

```python
xticks = np.arange(0,len(depth),2)
xlbl = [misc.round_sig_str(d) for d in depth[::2]*1e-6]
```

```python
yticks=np.arange(0,len(sep),2)
ylbl = [misc.round_sig_str(s) for s in sep[::2]]
```

```python
def dec(ax):
    ax.set_xticks( xticks );
    ax.set_xticklabels( xlbl );
    ax.set_yticks( yticks );
    ax.set_yticklabels( ylbl );
    ax.set_xlabel('Depth (MHz)')
    ax.set_ylabel(r'Separation $(w_0)$');
```

```python
fig, axs = plt.subplots(1,3, figsize=(22,5))
im = axs[0].imshow(simU)
plt.colorbar(im,ax=axs[0])
axs[0].set_title('U')
dec(axs[0])
im = axs[1].imshow(abs(mat['JMatrix']))#, vmax=1, vmin=0)
plt.colorbar(im,ax=axs[1])
axs[1].set_title('J')
dec(axs[1])
im = axs[2].imshow(UoJ)
plt.colorbar(im,ax=axs[2])
axs[2].set_title('U/J');
dec(axs[2])
```
