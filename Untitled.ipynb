```python
import matplotlib.pyplot as plt
import matplotlib.style
matplotlib.style.use('dark_background')
```

```python
import numpy as np
import scipy.io
```

```python
import Miscellaneous as misc
```

```python
mat = scipy.io.loadmat('MarksData.mat')
simU = mat['UMatrix']
simJ = mat['JMatrix']
depth = mat['potential'][:,0]
sep = mat['separation'][:,0]
```

```python
JoU = np.zeros(simU.shape)
```

```python
for i, ur in enumerate(simU):
    for j, u in enumerate(ur):
        if simJ[i,j] == 0:
            JoU[i,j] = None
        else:
            JoU[i,j] = abs(simJ[i,j]) / abs(simU[i,j])
```

```python
import scipy.optimize as opt
```

```python
drange = max(depth) - min(depth)
nump = len(depth)
mindepth = min(depth) - drange / (2*nump)
maxdepth = max(depth) + drange / (2*nump)

srange = max(sep) - min(sep)
nump = len(sep)
minsep = min(sep) - srange / (2*nump)
maxsep = max(sep) + srange / (2*nump)

ext = [mindepth/(20e6), maxdepth/(20e6), maxsep, minsep]
print(ext)
```

```python
ext
```

```python
intX, intY = [[], []]
for rnum, row in enumerate(JoU):
    func = lambda x: abs(np.interp(x, depth, row) - 0.12)
    res = opt.minimize(func, 0.2*1e6, tol=1e-6, method='Nelder-Mead')
    intX.append(res.x)
    intY.append(sep[rnum])
```

```python
def dec(ax):
    ax.set_xticks(depth[::tickSpacing]/(20e6))
    ax.set_yticks(sep[::tickSpacing])
    ax.set_xlabel('Depth (~mK)')
    ax.set_ylabel('Separation ($w_0$)')
```

```python
fig, axs = plt.subplots(1,3, figsize=(22,5))
asp= 2.03
im = axs[0].imshow(simU, extent=ext, aspect=asp)
tickSpacing = 5
plt.colorbar(im,ax=axs[0])
axs[0].set_title(r'$U$')
dec(axs[0])
im = axs[1].imshow(abs(mat['JMatrix']), extent=ext, aspect=asp)#, vmax=1, vmin=0)
plt.colorbar(im,ax=axs[1])
axs[1].set_title(r'$J$')
dec(axs[1])
cmapRange=0.12
im = axs[2].imshow(np.log(JoU), cmap='inferno', extent=ext, aspect=asp)
#im = axs[2].imshow(JoU, cmap='inferno', extent=ext, aspect=5e5)
plt.colorbar(im,ax=axs[2])
axs[2].set_title(r'$\log\{ J/U \}$');
lgood=50
axs[2].plot(np.array(intX[-lgood:])/(20e6), intY[-lgood:],'w:', label='J/U=0.12')
plt.legend()
dec(axs[2])
```
